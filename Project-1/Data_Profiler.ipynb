1. What is Data Analysis?

Data Analysis is the process of collecting, cleaning, transforming, and interpreting data to extract meaningful insights and support decision-making. It helps organizations identify patterns, trends, and relationships within data, enabling them to make informed business decisions. Data analysis is widely used in various fields such as business, healthcare, finance, and technology.

2. Data Science Project Lifecycle

A Data Science Project follows a structured lifecycle to solve business problems effectively. The main steps are:

- Problem Definition – Understand the business problem and objectives.
- Data Collection – Gather data from different sources such as CSV files, databases, APIs, and JSON files.
- Data Cleaning – Handle missing values, duplicates, and inconsistencies.
- Exploratory Data Analysis (EDA) – Analyze and visualize data to discover patterns and trends.
- Feature Engineering – Create and transform variables that improve model performance.
- Model Building – Train machine learning models using the prepared dataset.
- Model Evaluation – Measure model performance using evaluation metrics.
- Deployment – Deploy the model into a real-world environment.
- Monitoring and Maintenance – Continuously monitor and improve the model.


3. Machine Learning Problem Statement
Title:
Customer Churn Prediction Using Purchase Behavior Data

Problem Statement:
A consumer insights company wants to identify customers who are likely to stop purchasing products or services. The dataset contains customer information collected from multiple sources, including CSV files, JSON files, SQL databases, and APIs.

The objective is to build a machine learning classification model that predicts whether a customer will churn based on factors such as age, gender, income, purchase history, and spending behavior.

The target variable is:

Churn = Yes (Customer is likely to leave)
Churn = No (Customer is likely to stay)

This prediction will help the company improve customer retention and develop targeted marketing strategies.

4. What are Tensors?

A tensor is a mathematical data structure used to represent numerical data in machine learning and deep learning. Tensors are generalizations of scalars, vectors, and matrices into higher dimensions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("customers.csv")

print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.dtypes)

```

```python
df = pd.read_csv("customers.csv")

fig, axes = plt.subplots(1, 3, figsize=(15,5))

sns.countplot(x="Churn", data=df, ax=axes[0])
sns.countplot(x="Gender", data=df, ax=axes[1])
sns.histplot(df["Age"], kde=True, ax=axes[2])

plt.show()

print(df['Age'].describe())
print(f"Skewness: {df['Age'].skew():.2f}")
print(f"Kurtosis: {df['Age'].kurtosis():.2f}")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.histplot(df['Age'], ax=axes[0,0])
sns.boxplot(x=df['Age'], ax=axes[0,1],showfliers=False)
sns.countplot(x='Gender', data=df, ax=axes[1,0])
sns.countplot(x='Churn', data=df, ax=axes[1,1])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(df["Age"], kde=True, ax=axes[0])
axes[0].set_title("Age Distribution")

sns.histplot(df["AnnualIncome"], kde=True, ax=axes[1])
axes[1].set_title("Annual Income Distribution")

sns.histplot(df["AveragePurchaseAmount"], kde=True, ax=axes[2])
axes[2].set_title("Purchase Amount Distribution")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 5))

sns.barplot(
    data=df,
    x="Gender",
    y="AveragePurchaseAmount",
    ax=axes[0]
)
axes[0].set_title("Gender vs Purchase Amount")

sns.boxplot(
    data=df,
    x="Churn",
    y="AnnualIncome",
    ax=axes[1]
)
axes[1].set_title("Income vs Churn")

sns.violinplot(
    data=df,
    x="Churn",
    y="AnnualIncome",
    hue="Churn",
    inner="box"
)
axes[2].set_title("Annual Income Distribution by Churn")

plt.title("Annual Income Distribution by Churn")
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.select_dtypes(include="number").corr(),
    annot=True,
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(
    df,
    hue="Churn"
)

plt.show()
```
